```python
import numpy as np
import pandas as pd
```

```python
advertising = pd.read_csv("advertising.csv")
```

```python
advertising=(advertising - advertising.mean())/advertising.std()
```

```python
advertising.head()
```

```python
X = advertising["TV"]
Y = advertising["Sales"]
```

```python
# Conventional way to import seaborn
import seaborn as sns

# To visualise in the notebook
%matplotlib inline
```

```python
sns.pairplot(advertising, x_vars='Sales', y_vars='TV',size=7, aspect=0.7, kind='scatter')
```

```python
def gradient(X,Y,m_current,c_current,learning_rate=0.001,iters=1000):
    gd_df = pd.DataFrame(columns = ["m_current","c_current","cost"])
    for i in range(iters):
        y_current = m_current * X + c_current
        cost = sum([data**2 for data in (Y-y_current)])
        m_grad = 2* (sum((Y-y_current)*X))
        c_grad = 2* (sum(Y-y_current))
        gd_df.loc[i]=[m_current,c_current,cost]
        m_current = m_current + (m_grad*learning_rate)
        c_current = c_current + (c_grad*learning_rate)
    return(gd_df)
```

```python
X = np.array(X)
Y = np.array(Y)
```

```python
gradient(X,Y,0,0)
```

```python
gradient(X,Y,100,100)
```
