```python
import pandas as pd
import numpy as np
```

```python
car_price= pd.read_csv("./CarPrice_Assignment.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
car_price
```

```python
car_price.info()
okay = ['fueltype', 'aspiration', 'doornumber', 'carbody', 'drivewheel', 'enginelocation', 'enginetype', 'cylindernumber', 
       'fuelsystem', ]
numeric_cols = ['symboling','wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight', 'enginesize', 'boreratio', 'stroke', 
               'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg' ,'price']
```

```python
for column in okay:
    status = pd.get_dummies(car_price[column], drop_first = True)
    car_price = pd.concat([car_price, status], axis = 1)
    car_price.drop([column], axis = 1, inplace = True)
```

```python
#we also drop CarName and car_ID
car_price.drop(['CarName'], axis =1, inplace = True)

car_price.head()
```

```python
# so far we have cleaned the data
```

```python
# for efficiency reasons and to make analysis alot easier, we will now scale the data, i.e. make it such that 
# the values stored are actually deviation from the mean.
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
# we only need to scale the columns which have actual numeric values
car_price[numeric_cols] = scaler.fit_transform(car_price[numeric_cols])
car_price.head()
```

```python
# we now split the data
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
df_train, df_test = train_test_split(car_price, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
# to make it into X and y
y_train = df_train.pop('price')
X_train = df_train
```

```python
# FINALLY TO FIND THE REGRESSION
```

```python
import statsmodels.api as sm
X_train_lm = sm.add_constant(X_train)

lr_1 = sm.OLS(y_train, X_train_lm).fit()

lr_1.params
```

```python
print(lr_1.summary())
```

```python
# NOW, to see which columns are redundant, we see the ones with high P-value and high VIF
# to see the VIF
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
# to drop a few values
X = X_train.drop('mfi', 1,)
X = X_train.drop('spfi', 1,)
```

```python
# Build a new model with the dropped stuff and see again, and drop again, and so on
X_train_lm = sm.add_constant(X)

lr_2 = sm.OLS(y_train, X_train_lm).fit()
```

```python
print(lr_2.summary())
```

```python
vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
X = X.drop('mfi', 1,)
X_train_lm = sm.add_constant(X)

lr_2 = sm.OLS(y_train, X_train_lm).fit()
```

```python
print(lr_2.summary())
```

```python
vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
X = X.drop('rotor', 1,)
X = X.drop('two', 1,)
X = X.drop('l', 1,)

X_train_lm = sm.add_constant(X)

lr_2 = sm.OLS(y_train, X_train_lm).fit()
```

```python
print(lr_2.summary())
```

```python
vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
X = X.drop('carheight', 1,)


X_train_lm = sm.add_constant(X)

lr_2 = sm.OLS(y_train, X_train_lm).fit()
```

```python
print(lr_2.summary())
```

```python
vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
# so we have eliminated some clumns based on p-values and VIF
```

```python
# now we do reidual analysis and see if it satisfies the requirments of linear regression,
```

```python
# first we check if the error terms are normally distributed
```

```python
y_train_pred = lr_2.predict(X_train_lm)
```

```python
fig = plt.figure()
sns.distplot((y_train - y_train_pred), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)     
```

```python
# this looks good
```

```python
# NOW we prepae the test datas sets, and then evaluate our model
```

```python
y_test = df_test.pop('price')
X_test = df_test
```

```python
X_test_sm = sm.add_constant(X_test)
```

```python
X_test_sm = X_test_sm.drop(['carheight', 'rotor', 'two', 'l' , 'mfi' , 'spfi' ], axis = 1)
```

```python
y_test_pred= lr_2.predict(X_test_sm)
```
