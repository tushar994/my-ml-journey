First we will convert the columns with numbers to type int

```python
data_set['Total Ballots Counted (Estimate)'].sum()
```

```python
number_columns = ['Total Ballots Counted (Estimate)','Vote for Highest Office (President)', 'Voting-Eligible Population (VEP)', 'Voting-Age Population (VAP)', 'Prison', 'Probation', 'Parole',  'Total Ineligible Felon',  'Overseas Eligible']

## remove commas from the strings created
data_set[number_columns] = data_set[number_columns].replace(',','', regex=True)
# print(data_set)
## set nan to NaN
data_set[number_columns] = data_set[number_columns].replace('nan','NaN', regex=True)
# print(data_set)
## convert to float64
data_set[number_columns] = data_set[number_columns].astype('float64')
data_set
```

```python
data_set['Total Ballots Counted (Estimate)'].sum()
```

```python
data_set.dtypes
```

```python
number_columns = ['VEP Turnout Rate', '% Non-citizen']
## remove commas from the strings created
data_set[number_columns] = data_set[number_columns].replace('%','', regex=True)
# print(data_set)
## set nan to NaN
data_set[number_columns] = data_set[number_columns].replace('nan','NaN', regex=True)
# print(data_set)
## convert to float64
data_set[number_columns] = data_set[number_columns].astype('float64')
data_set
```

We have converted the numeric columns into numbers, now we will fix the rows

The steps to fix the rows are
- delete summary rows
- delete incorrect rows - like unnecessary header rows, or footer rows
- delete extra rows

And to fix the columns, we have to 
- rename column names to make more convenient
- delete columns that are not needed
- Split columns such that you can get more data (for example)
    - date can be split into year, month, day
    - address can be split
    - From a name a title can be extracted
- merge some columns
    - name from last name and first name
    - state district from state and district
    - This is mostly done to create more unique values in one column, over having repetition in two or more columns

We will not be deleting any rows/columns now, as that depends on the use case of the data, and we have no use case for this data.

Now To deal with missing values. What can we do here? In this particular data frame, i will leave the NaN values as NaN, as using a median, or average desnt make much sense to me, as they are values that pertain to particular states, and in this case, the value of these columns for each state is independant of the values for the other states. All the other steps do not need to be done for this one (besides standardisation of the official/unoffical column), hopefully in other datasets we'll get to do more.

```python
# 'Official/Unofficial'
data_set['Official/Unofficial'] = data_set['Official/Unofficial'].str.capitalize()
data_set
```
